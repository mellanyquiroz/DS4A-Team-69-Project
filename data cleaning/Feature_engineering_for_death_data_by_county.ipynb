```python
# import packages
import pandas as pd
import numpy as np
```

```python
# load file
df = pd.read_csv("/home/thuy/Documents/Data Science/02. DS4A/Capstone_teamrepo/DS4A-TeamProject/data/Pre-merging/Opioid_Deaths_by_County_2015-2020.csv")
df.head()
```

```python
# remove California total number and Unknown
df = df[df["County"] != "California"]
df = df[df["County"] != "Unknown"]
df["County"].unique()
```

```python
# check data types
df.info()
```

```python
# make sure County names don't have space (prepare for joining)
df["County"] = df["County"].str.strip()
```

```python
# divide the rates and counts into precovid and covid
precovid_df = df[df["Year"] != 2020]
precovid_df["Year"].value_counts()
```

```python
precovid_county = precovid_df.groupby("County")[["Rates","Counts"]].mean()
precovid_county
```

```python
# During COVID
covid_county = df[df["Year"] == 2020]
covid_county
```

```python
# Join 2 dataframes
county_df = precovid_county.join(covid_county.set_index("County"),lsuffix="_precovid", rsuffix="_covid")
```

```python
county_df = county_df.drop(columns="Year")
county_df.head()
```

```python
# Rename columns
county_df = county_df.rename(columns=
{"Rates_precovid":"PRECOVID_RATES",
"Counts_precovid":"PRECOVID_COUNTS",
"Rates_covid":"COVID_RATES",
"Counts_covid":"COVID_COUNTS"})
county_df.head()
```

```python
# load hospital-related metrics
facility_df = pd.read_csv("/home/thuy/Documents/Data Science/02. DS4A/Capstone_teamrepo/DS4A-TeamProject/data/Merged_deaths_hospital_availability_by_county.csv")
facility_df.head()
```

```python
# join with pre and during covid data
county_merged = county_df.join(facility_df.set_index("COUNTY"))
```

```python
county_merged.head()
```

```python
# remove unnecessary columns
county_merged = county_merged.drop(columns=["Unnamed: 0","DEATHRATES19","DEATHCOUNTS19","DEATHRATES20","DEATHCOUNTS20","COUNTS_GR19","COUNTS_GR20"])
county_merged.head()
```

```python
county_merged = county_merged.reset_index().rename(columns={"County":"COUNTY"})
```

```python
county_merged = county_merged.fillna(0)
county_merged
```

```python
# save to csv
county_merged.set_index("COUNTY").to_csv("/home/thuy/Documents/Data Science/02. DS4A/Capstone_teamrepo/DS4A-TeamProject/data/Merged_precovid_duringcovid_deaths_and_hospital_metrics.csv")
```
